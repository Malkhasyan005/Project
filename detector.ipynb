```python
import pandas as pd
import numpy as np
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
fake_data = pd.read_csv('news/Fake.csv')
true_data = pd.read_csv('news/True.csv')
```

```python
true_data.head()
```

```python
fake_data.head()
```

```python
true_data['label'] = 1
fake_data['label'] = 0
```

```python
true_data.head()
```

```python
news = pd.concat([true_data, fake_data], axis=0)
news.head()
```

```python
news.tail()
```

```python
news.isnull().sum()
```

```python
news = news.drop(columns=['date', 'title', 'subject'], axis=1)
```

```python
news.head()
```

```python
news = news.sample(frac=1)
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news.head()
```

```python
news.drop(['index'], axis=1, inplace=True)
```

```python
news.head()
```

Working with texts

```python
def wordopt(text):
    text = text.lower()
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'\d', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\n', ' ', text)
    return text
```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news.head()
```

# Training and Testing

```python
x = news['text']
y = news['label']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
vectorization = TfidfVectorizer()
```

```python
xv_train = vectorization.fit_transform(x_train)
```

```python
xv_test = vectorization.transform(x_test)
```

```python
LR = LogisticRegression()
```

```python
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
DT = DecisionTreeClassifier()
```

```python
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(xv_train, y_train)
```

```python
pred_rfc = rfc.predict(xv_test)
```

```python
rfc.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rfc))
```

```python
gbc = GradientBoostingClassifier()
```

```python
gbc.fit(xv_train, y_train)
```

```python
gbc_pred = gbc.predict(xv_test)
```

```python
gbc.score(xv_test, y_test)
```

```python
print(classification_report(y_test, gbc_pred))
```

```python
def output(pred):
    if pred == 1:
        return 'True News'
    else:
        return 'Fake News'
```

```python
def manual_testing(news):
    testing_news = {'text': [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test['text'] = new_def_test['text'].apply(wordopt)
    xv_new = vectorization.transform(new_def_test['text'])
    pred_lr = LR.predict(xv_new)
    pred_gbc = gbc.predict(xv_new)
    pred_rfc = rfc.predict(xv_new)
    return f'\n\nLR Prediction: {output(pred_lr[0])} \nGBC Prediction: {output(pred_gbc[0])} \nRFC Prediction: {output(pred_rfc[0])}'
```

```python
news_input = input("Enter the news text to classify: ")
```

```python
print(manual_testing(news_input))
```
